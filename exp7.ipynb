```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
### Create the dataset
```

```python
# week(independent variable)
X=np.array([1,2,3,4,5]).reshape(-1,1)
#sales in thousands (depeneant variable)
y=np.array([1.2,1.8,2.6,3.2,3.8])
```

```python
X
```

```python
y
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
#y=a +bx
#Intercept(a)
a=model.intercept_
#slope(b)
```

```python
b= model.coef_[0]

print("intercept (a):",a)
print("slope (b):",b)
```

```python
#y=0.54+0.66x
```

```python
plt.scatter(X,y,label="Actual Data")
plt.plot(X,model.predict(X),label="Regression line")
plt.xlabel("week");
plt.ylabel("sales in (thousands)")
plt.title("linear Regression : week vs sales")
plt.legend()
plt.show()
```
