```python
# imports
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

n = 200

USD = np.random.uniform(80, 90, n)
Crude = np.random.uniform(65, 85, n)
Nifty = np.random.uniform(17500, 19000, n)

Gold = (
    USD * 300 +
    Crude * 50 +
    Nifty * 1.5 +
    np.random.normal(0, 500, n)
)

df = pd.DataFrame({
    'USD': USD,
    'Crude': Crude,
    'Nifty': Nifty,
    'Gold': Gold
})

df = df.round(2)
df.head()
```

```python
X = df[['USD', 'Crude', 'Nifty']]
y = df['Gold']

model = LinearRegression()
model.fit(X, y)
```

```python
prediction = model.predict([[14, 74, 18600]])
print("Predicted Gold Price:", prediction.round(2)[0])
```

```python
plt.scatter(y, model.predict(X))
plt.xlabel("Actual Gold Price")
plt.ylabel("Predicted Gold Price")
plt.title("Actual vs Predicted Gold Price")
plt.plot(y, y, linestyle='--')
plt.show()
```
